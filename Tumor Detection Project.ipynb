This notebook aims to build a machine learning model to classiy malignant and benign tumors. 

```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
data = pd.read_csv('data/Tumor_Detection.csv')

# Displaying the first few rows of the dataset
print(data.shape)
data.head()
```

```python
# Displaying column names of the dataset
data.columns
```

```python
# Displaying basic information about the dataset
data.info()
```

```python
# Dropping the 'id' column as it is not needed for analysis
data.drop(['id'], axis=1, inplace=True)

# Start points
l = list(data.columns)
features_mean = l[1:11]
features_se = l[11:21]
features_worst = l[21:31]

# Displaying the features
print("Features Mean:", features_mean)
print("Features SE:", features_se)
print("Features Worst:", features_worst)
```

```python
# Visualizing the distribution of tumor types
sns.countplot(x='diagnosis', data=data, palette='Set1')
plt.title('Count of Tumor Types')
plt.xlabel('Tumor Type')
plt.ylabel('Count')
plt.show()
```

```python
# Displaying basic statistics of the dataset
data.describe()
```

```python
# Displaying the correlation matrix of the dataset
data.corr(numeric_only=True)
```

```python
# Visualizing the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(numeric_only=True), cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Mapping the diagnosis column to numerical values
data['diagnosis'] = data['diagnosis'].map({'M': 1, 'B': 0})

# Displaying the updated dataset
data.head()
```

```python
# Splitting the dataset into features and target variable
X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis']

# Displaying the shape of features and target variable
print("Features shape:", X.shape)
print("Target shape:", y.shape)
```

```python
# Splitting the dataset into training and testing sets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print("Training set shape:", X_train.shape, y_train.shape)
print("Testing set shape:", X_test.shape, y_test.shape)
```

```python
# Standardizing the features
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)

X_train
```

```python
# Training a Random Forest Classifier
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)

# Making predictions on the test set
y_pred = rf.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
# Evaluating classification report
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
